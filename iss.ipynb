```python
import json
import requests
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time

lat = []
long = []

for i in range(100):
    data = requests.get('http://api.open-notify.org/iss-now.json')
    data_results = data.text
    info = json.loads(data_results)
    lat.append(info['iss_position']['latitude'])
    long.append(info['iss_position']['longitude'])
    time.sleep(10)

print(lat)
print(long)
```

```python
plt.plot(long, lat, label = 'location')
plt.title('International Space Station')
plt.xlabel('longitude')
plt.ylabel('latitude')

plt.xticks([-15,0,15,45,75,105])
plt.yticks([-20,0,20,40,60,80,100])

plt.legend()
plt.show()
```
