```python
import json
import requests
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time

lat = []
long = []

for i in range(100):
    data = requests.get('http://api.open-notify.org/iss-now.json')
    info = data.json()
    
    lati = float(info['iss_position']['latitude'])
    longi = float(info['iss_position']['longitude'])
    
    lat.append(lati)
    long.append(longi)
       
    time.sleep(10)
```

```python
plt.plot(long, lat, label = 'location')
plt.title('Location of the International Space Station')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.legend()
plt.show()
```
